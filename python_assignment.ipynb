```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
import random
```

```python
df['Height'] = np.random.randint(150,181,size = df.shape[0])
```

```python
df.head(10)
```

```python
# Analysis Tasks:
```

```python
team_distribution = df['Team'].value_counts(normalize=True)*100
```

```python
print("Distribution of employees across each team and  the percentage:\n", team_distribution)
```

```python
position_distribution = df['Position'].value_counts()
```

```python
print("Segregation by Positions:\n",position_distribution)
```

```python
age_distribution = df['Age'].value_counts()
```

```python
age_distribution
```

```python
age_bins = [20,30,40,50,60,70]
df['age_group'] = pd.cut(df['Age'], bins=age_bins)
predominant_age_group = df['age_group'].value_counts().idxmax()
```

```python
print("Predominant Age Group is: ",predominant_age_group)
```

```python
df.head(20)
```

```python
salary_expenditure = df.groupby(['Team','Position'])['Salary'].sum().reset_index()  
```

```python
max_expenditure = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]  
```

```python
print("Highest Salary Expenditure by Team and Position:",max_expenditure)
```

```python
correlation = df['Age'].corr(df['Salary'])
```

```python
correlation
```

```python
plt.scatter(df['Age'],df['Salary'],marker ='p',edgecolors='r')
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')  
plt.ylabel('Salary')  
plt.show()
```

```python
# Graphical Representation:
```

1.distribution of employees across each team

```python
team_counts = df['Team'].value_counts()
team_names = team_counts.index
```

```python
y = team_distribution
labels = team_names
plt.figure(figsize=(15,43))
plt.pie(y,labels = labels)
plt.style.use('ggplot')
plt.title("Employee Distribution Across Teams")
plt.show()
```

2.Segregation by po

```python
x = position_distribution.index
y = position_distribution
sns.set_palette("Set2")
plt.bar(x,y)
plt.title("Segregation by Positions")
plt.xlabel("Positions")
plt.ylabel("Count of Employees")
```

3.  Predominant age group

```python
x=df['Age']
plt.hist(x,color='olive')
plt.title("Predominant Age Group")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

4. Highest Salary Expenditure:

```python
salary_pivot = salary_expenditure.pivot(index='Team', columns='Position', values='Salary')

plt.figure(figsize=(12, 8))  
sns.heatmap(salary_pivot, annot=True, fmt=".0f", cmap="YlGnBu", cbar_kws={'label': 'Total Salary ($)'})  
plt.title('Salary Expenditure by Team and Position')  
plt.xlabel('Position')  
plt.ylabel('Team')  
plt.tight_layout()  
plt.show()
```

5. Relationship Between Age and Salary

```python
x = df["Age"]
y=df["Salary"]
plt.title('Relationship Between Age and Salary')
sns.scatterplot(x=x,y=y,hue=df["Salary"])
plt.show()
```
